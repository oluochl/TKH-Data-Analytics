## Why this order

The sequence is not arbitrary. Each decision was made to ensure that when you encounter a new concept, you already have what you need to understand it.

**Tools first.** You cannot do anything without a working environment. Python installed, Jupyter running, VS Code configured — this is table stakes. It goes first because everything else depends on it.

**Python before statistics.** You need the language before you can apply the concepts. A student who does not know pandas cannot do meaningful exploratory data analysis, no matter how well they understand the theory.

**Statistics before math.** Statistical intuition comes before mathematical formalism. Understanding that a p-value is about surprise — not proof — is more important than knowing the formula. We build the intuition first, then give it mathematical grounding.

**Math before ML.** When you know what a dot product means geometrically, linear regression makes sense. When you understand a gradient, gradient descent is obvious. The math notebooks exist so that when you see the algorithm, you understand why it works — not just how to use it.

**EDA before modeling.** You cannot make good modeling decisions about data you have not looked at. Every experienced data scientist has a story about a model that failed because of something that would have been obvious in EDA. We do EDA first because it should always come first.

**Preprocessing before algorithms.** Garbage in, garbage out. A perfect algorithm fed bad data produces bad results. You need to know how to prepare data before you learn the algorithms that consume it.

**ML concepts before algorithms.** Understanding bias and variance, overfitting, cross-validation, and evaluation metrics before you touch a specific algorithm means you can reason about every algorithm you will ever encounter. The concepts notebook is the lens through which all the algorithm notebooks should be read.

**Supervised before unsupervised.** Learning with labels is conceptually simpler than learning without them. We establish the framework with supervised learning, then extend it to the harder problem of unsupervised learning.

## What you will be able to do after each folder

Here is what gets unlocked at each stage of the curriculum:

**After `tools/`:** Set up a professional data science development environment from scratch on any machine.

**After `python/`:** Write data manipulation scripts from scratch using pandas and numpy. Load, clean, filter, group, and summarize datasets programmatically.

**After `statistics/`:** Interpret any statistical result you encounter — means, medians, standard deviations, confidence intervals, p-values, and correlation coefficients — in plain English.

**After `math/`:** Understand why machine learning algorithms work at a mathematical level. Recognize what optimization is doing and why it matters.

**After `eda/`:** Explore any new dataset with confidence. Know which questions to ask, which visualizations to create, and what patterns to look for before you touch a model.

**After `preprocessing/`:** Prepare any real-world dataset for machine learning. Handle missing values, encode categories, scale features, and split data without introducing data leakage.

**After `ml_concepts/`:** Think like a machine learning practitioner. Reason about overfitting, model selection, and evaluation before choosing an algorithm.

**After `feature_engineering/`:** Extract more signal from any dataset. Create features that help models learn what raw columns cannot express.

**After `supervised/`:** Build, tune, and evaluate twelve different machine learning algorithms. Know when to use each one and when not to.

**After `unsupervised/`:** Find structure in data without labels — cluster customers, reduce dimensions, discover patterns that no one thought to look for.

**After `model_evaluation/`:** Know whether your model is actually good. Use the right metrics, diagnose failure modes, and communicate uncertainty honestly.

**After `case_studies/`:** Complete an end-to-end data science project from raw data to final recommendation. Do it on your own, on a new domain, with no scaffolding.

## A final note before you begin

What you are about to do is genuinely difficult. Data science sits at the intersection of statistics, programming, and domain knowledge — three disciplines that each take years to master individually. You are learning all three simultaneously, in a compressed timeline, while also building the habits and workflows of a professional.

There will be days when nothing makes sense. There will be labs where you feel completely lost. There will be moments where you look at someone else's notebook and think "I could never write that."

That is normal. That is what learning looks like from the inside.

The students who succeed are not the ones who never struggle. They are the ones who keep coming back  and who run the cell one more time, ask the question they were embarrassed to ask, and return to the notebook they did not understand the first time.
